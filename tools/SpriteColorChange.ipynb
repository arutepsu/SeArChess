```python
import cv2
import numpy as np
import os

input_folder = r"C:\SeArChess\src\main\resources\assets\images\bishop"

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".png"):
        
        path = os.path.join(input_folder, filename)

        # Bild mit Alpha laden
        image = cv2.imread(path, cv2.IMREAD_UNCHANGED)

        if image.shape[2] == 4:
            bgr = image[:, :, :3]
            alpha = image[:, :, 3]
        else:
            bgr = image
            alpha = None

        # HSV
        hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)

        # Blau erkennen
        lower_blue = np.array([100, 150, 50])
        upper_blue = np.array([140, 255, 255])
        mask = cv2.inRange(hsv, lower_blue, upper_blue)

        # Blau → Rot
        bgr[mask != 0] = [0, 0, 255]

        # Alpha zurück
        if alpha is not None:
            result = cv2.merge([bgr[:, :, 0], bgr[:, :, 1], bgr[:, :, 2], alpha])
        else:
            result = bgr

        # 🔥 NEU: Name ändern black → white
        if filename.startswith("black_"):
            new_filename = filename.replace("black_", "white_", 1)
        else:
            new_filename = filename  # falls kein "black_" drin ist

        output_path = os.path.join(input_folder, new_filename)

        cv2.imwrite(output_path, result)

print("Fertig ✅")
```

```python
import cv2
import numpy as np
import os

input_folder = r"C:\SeArChess\src\main\resources\assets\images\rook"

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".png"):
        
        path = os.path.join(input_folder, filename)

        # Bild mit Alpha laden
        image = cv2.imread(path, cv2.IMREAD_UNCHANGED)

        if image.shape[2] == 4:
            bgr = image[:, :, :3]
            alpha = image[:, :, 3]
        else:
            bgr = image
            alpha = None

        # In HSV umwandeln
        hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)

        # Helligkeit und Sättigung extrahieren
        h, s, v = cv2.split(hsv)

        # Maske: helle & wenig gesättigte Pixel (Weiß/Grau)
        mask = (v > 180) & (s < 40)  # Werte ggf. anpassen

        # Hue auf Rot setzen
        h[mask] = 0  # Rot

        # Neue HSV zusammenbauen
        hsv_new = cv2.merge([h, s, v])

        # Zurück zu BGR
        bgr = cv2.cvtColor(hsv_new, cv2.COLOR_HSV2BGR)

        # Alpha wieder anhängen
        if alpha is not None:
            result = cv2.merge([bgr[:, :, 0], bgr[:, :, 1], bgr[:, :, 2], alpha])
        else:
            result = bgr

        # Name ändern: black → white
        if filename.startswith("black_"):
            new_filename = filename.replace("black_", "white_", 1)
        else:
            new_filename = filename

        output_path = os.path.join(input_folder, new_filename)

        cv2.imwrite(output_path, result)

print("Fertig ✅")
```

```python
import cv2
import numpy as np
import os

input_folder = r"C:\SeArChess\src\main\resources\assets\images\rook"

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".png"):
        
        path = os.path.join(input_folder, filename)

        image = cv2.imread(path, cv2.IMREAD_UNCHANGED)

        if image.shape[2] == 4:
            bgr = image[:, :, :3]
            alpha = image[:, :, 3]
        else:
            bgr = image
            alpha = None

        hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)

        # 🔴 Rot erkennen
        lower_red1 = np.array([0, 120, 70])
        upper_red1 = np.array([10, 255, 255])
        lower_red2 = np.array([170, 120, 70])
        upper_red2 = np.array([180, 255, 255])

        mask_red = cv2.inRange(hsv, lower_red1, upper_red1) + \
                   cv2.inRange(hsv, lower_red2, upper_red2)

        # ⚪ Weiß/Grau erkennen
        h, s, v = cv2.split(hsv)
        mask_white = ((s < 40) & (v > 180)).astype(np.uint8) * 255

        # 🔥 Kombinieren
        mask = mask_red + mask_white

        # 🔵 Alles → Blau
        bgr[mask != 0] = [255, 0, 0]

        # Alpha zurück
        if alpha is not None:
            result = cv2.merge([bgr[:, :, 0], bgr[:, :, 1], bgr[:, :, 2], alpha])
        else:
            result = bgr

        # Name ändern
        if filename.startswith("white_"):
            new_filename = filename.replace("white_", "black_", 1)
        else:
            new_filename = filename

        output_path = os.path.join(input_folder, new_filename)

        cv2.imwrite(output_path, result)

print("Fertig ✅")
```
